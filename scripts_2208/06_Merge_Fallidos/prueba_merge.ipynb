```python
import pandas as pd

def merge_particles(data_dict, particle_type):
    merged_dict = {}
    
    # Assuming alphas are 4, 5, and 6 as per your example
    alphas = [4, 5, 6]
    types = ['TTH', 'WH', 'ZH']

    for alpha in alphas:
        # List to hold DataFrames of the same alpha for merging
        dfs = []
        
        for type in types:
            if alpha in data_dict[type]:
                dfs.append(data_dict[type][alpha][particle_type])
        
        # Concatenate all DataFrames in the list
        if dfs:
            merged_df = pd.concat(dfs, ignore_index=True)
            merged_dict[alpha] = merged_df
    
    return merged_dict

# Example usage
data_dict = {
    'TTH': {
        4: {'muons': pd.DataFrame({'data': [1]}), 'electrons': pd.DataFrame({'data': [2]})},
        5: {'muons': pd.DataFrame({'data': [3]}), 'electrons': pd.DataFrame({'data': [4]})},
        6: {'muons': pd.DataFrame({'data': [5]}), 'electrons': pd.DataFrame({'data': [6]})},
    },
    'WH': {
        4: {'muons': pd.DataFrame({'data': [7]}), 'electrons': pd.DataFrame({'data': [8]})},
        5: {'muons': pd.DataFrame({'data': [9]}), 'electrons': pd.DataFrame({'data': [10]})},
        6: {'muons': pd.DataFrame({'data': [11]}), 'electrons': pd.DataFrame({'data': [12]})},
    },
    'ZH': {
        4: {'muons': pd.DataFrame({'data': [13]}), 'electrons': pd.DataFrame({'data': [14]})},
        5: {'muons': pd.DataFrame({'data': [15]}), 'electrons': pd.DataFrame({'data': [16]})},
        6: {'muons': pd.DataFrame({'data': [17]}), 'electrons': pd.DataFrame({'data': [18]})},
    }
}

# Merge for electrons and muons separately
merged_electrons = merge_particles(data_dict, 'electrons')
merged_muons = merge_particles(data_dict, 'muons')

# The merged dictionaries will contain the merged DataFrames for each alpha
print("Merged Electrons:")
for alpha, df in merged_electrons.items():
    print(f"Alpha {alpha}:")
    print(df)

print("\nMerged Muons:")
for alpha, df in merged_muons.items():
    print(f"Alpha {alpha}:")
    print(df)
```
